```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Carregar áudio
audio_path = "C:/Users/luqui/Downloads/duzeeeeentos.mp3"
y, sr = librosa.load(audio_path, sr=None)

# Estimar frequência fundamental (pitch) com YIN
f0 = librosa.yin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))

# Criar eixo do tempo para o gráfico
times = librosa.times_like(f0, sr=sr)

# Remover valores NaN da frequência fundamental
f0 = np.nan_to_num(f0, nan=np.mean(f0[np.isfinite(f0)]))

# Gerar sinal senoidal baseado na frequência fundamental
senoidal = np.sin(2 * np.pi * np.cumsum(f0) / sr)

# Plotar o sinal original
plt.figure(figsize=(10, 4))
plt.plot(times, y[:len(times)], label="Sinal Original", alpha=0.5)
plt.plot(times, senoidal[:len(times)], label="Sinal Senoidal", color="red")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Sinal Original vs. Sinal Senoidal")
plt.legend()
plt.grid()
plt.show()
```
